```python
import scrapy

class HoopshypeSpider(scrapy.Spider):
    name = 'hoopshype_spider'
    start_urls = ['https://hoopshype.com/salaries/players/']

    def parse(self, response):
        salaries = response.css('.name a::text').getall()
        players = response.css('.contract-value span::text').getall()

        yield salaries, players
```
